```python
from bs4 import BeautifulSoup
import requests
```

Sending get request to the webpage server to get the source code of the page

```python
page = requests.get('https://www.dineout.co.in/top-10-restaurants-in-kolkata')
```

```python
page
```

Page Content

```python
soup = BeautifulSoup(page.content)
soup
```

scrapping first name

```python
# first, we will use html tag where we have th first title of the restaurants.

first_title = soup.find('a',class_="restnt-name ellipsis")
first_title
```

```python
first_title.text
```

scrapping first location

```python
loc = soup.find('div',class_="restnt-loc ellipsis")
loc.text
```

scrapping first price

```python
sfp = soup.find('span',class_="double-line-ellipsis")
sfp.text.split("|")[0]
```

scrapping multiple titles

```python
# Now we have all the tags in which there are the titles.
# Now we will extract the text from these tags one by one by looping over these tags
titles = [] #empty list for storing
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    titles.append(i.text)
titles
```

```python
location = []
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location
```

```python
price = []
for i in soup.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text)
price
```

```python
images = []
for i in soup.find_all('img',class_="no-img"):
    images.append(i.get('data-src'))
images
```

```python
# printing length
```

```python
print(len(titles),len(location),len(price),len(images))
```

```python
# making dataframe
import pandas as pd
df = pd.DataFrame({'Titles':titles,'Location':location,'Price':price,'Image_url':images})
df
```
