```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
job_url=[]
```

```python
#scrapping job title from the given page
title_tags=driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)

#scrapping job location from the given page
location_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft fs12 lh16 locWdth"]')
for i in location_tags[0:10]:
    location=i.text
    job_location.append(location)

#scrapping company name from the given page
company_tags=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in company_tags[0:10]:
    name=i.text
    company_name.append(name)
#scrapping job experience from the given page
experience_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft fs12 lh16 expwdth"]')
for i in experience_tags[0:10]:
    experience=i.text
    experience_required.append(experience)

#scrapping job url from the given page
url_tags=driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
for i in url_tags[0:10]:
    url=i.get_attribute('href')
    job_url.append(url)
    
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required),len(job_url))
```

```python
#making dataframe

df=pd.DataFrame({'Job_title':job_title,'Job_location':job_location,'Company_name':company_name,'Experience_required':experience_required,'Job_url':job_url})
df
```

Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.

The above task will be done in following steps:

1.First get the webpage https://www.azquotes.com/
2. Click on Top Quotes
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
driver=webdriver.Chrome(r'C:\Users\nikhi\Downloads\Compressed\chromedriver.exe')
driver.get('https://www.azquotes.com/')
```

```python
#clicking on top quotes
quotes=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
quotes.click()
```

```python
quote=[]
author=[]
typp=[]
```

```python
start=0
end=10
for page in range(start,end):
    qutoes= driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in qutoes:
        quote.append(i.text)
        

    authors= driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in authors:
        author.append(i.text)
        

    types= driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in types:
        typp.append(i.text)
        
        
    next_button= driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]/a")
    next_button.click()
    time.sleep(3)
    
```

```python
print(len(quote),len(author),len(typp))
```

```python
#creating dataframe 
df=pd.DataFrame({'Quote':quote,'Author':author,'Type of Quotes':typp})
df
```
